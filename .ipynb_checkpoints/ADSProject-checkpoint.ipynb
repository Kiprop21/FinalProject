**Introduction**
This project aims to explore and visualize the clustering of regions in Nigeria based on annual rainfall and location. The project entails grouping sites into distinct clusters that can be used to inform the design of parametric crop insurance products.

**Step 1: Data Collection and Preparation**

*1.1 Loading the libraries*

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

*1.2 Loading the Dataset and Exploratory Data Analysis*

```python
data=pd.read_csv("D:/Desktop/Python/FinalProject/Nigeria_Rainfall_Data.csv")
data.head()
```

```python
# Display the first few rows of the data to understand its structure
data.columns
```

```python
# Melt the DataFrame to have a single 'Date' column and a 'Rainfall' column
# The columns to keep as identifiers are 'No.', 'District', 'PODs', 'Site_Name', 'N', 'S', 'W', 'E'
melted_data = pd.melt(data, 
                      id_vars=['No.', 'District', 'PODs', 'Site_Name', 'N', 'S', 'W', 'E'], 
                      var_name='Date', 
                      value_name='Rainfall')
```

```python
melted_data.head() 
```

```python
#Check for missing values
melted_data.isnull().sum()
```

```python
# Convert the "Date" column to datetime format
melted_data['Date'] = pd.to_datetime(melted_data['Date'], format='%m/%d/%Y')
melted_data.head()
```

```python
# Replace non-numeric rainfall values with NaN and convert the column to numeric
melted_data['Rainfall'] = pd.to_numeric(melted_data['Rainfall'], errors='coerce')
melted_data.head()
```

```python
# Extract the year from the "Date" column
melted_data['Year'] = melted_data['Date'].dt.year
melted_data.head()
```

```python
# Get the annual rainfall per site
annual_rainfall = melted_data.groupby(['No.', 'District', 'PODs', 'Site_Name', 'N', 'S', 'W', 'E', 'Year'], as_index=False)['Rainfall'].sum()
annual_rainfall.head()
```

```python
# Get the total rainfall per site for all the years
total_rainfall = annual_rainfall.groupby(['No.', 'District', 'PODs', 'Site_Name', 'N', 'S', 'W', 'E'], as_index=False)['Rainfall'].sum()
total_rainfall.head()
```

**Step 2: Data Pre Processing**

*2.1 Data Engineering*

```python
#Find the latitudes and longitudes
total_rainfall['latitude']=total_rainfall[['N','S']].mean(axis=1)
total_rainfall['longitude']=total_rainfall[['W','W']].mean(axis=1)
total_rainfall.head()
```

*2.2 Feature Scaling*

```python
scaler = StandardScaler()
total_rainfall[['lat_scaled','lon_scaled','Rainfall_Scaled']] = scaler.fit_transform(total_rainfall[['latitude','longitude','Rainfall']])
```

```python
total_rainfall.head()
```

*2.3 Feature Selection*

```python
data =total_rainfall[['lat_scaled','lon_scaled','Rainfall_Scaled']]
data.head()
```

**Step 5: Clustering**

```python
#Check for optimal number of clusters
inertia = []
K = range(1, 20)  # Test for 1 to 10 clusters

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Perform K-Means Clustering
kmeans = KMeans(n_clusters=7, random_state=42)  
clusters = kmeans.fit_predict(data)

# Attach cluster labels to the original dataframe
total_rainfall['Cluster'] = clusters

# Evaluate with silhouette score
score = silhouette_score(data, clusters)
print(f'Silhouette Score: {score}')
```

```python
total_rainfall.head()
```

```python
#Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='Cluster', data=total_rainfall, palette='viridis')
plt.title("Clusters Visualization")
plt.show()
```
